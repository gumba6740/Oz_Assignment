[문제 1]

사용자로부터 두 개의 숫자를 입력받아 나눗셈을 수행하는 프로그램을 작성하세요. 

[예외처리]
- 입력 값이 숫자가 아닐 때 발생할 수 있는 ValueError
- 나누는 수가 0일 때 발생할 수 있는 ZeroDivisionError
- 위 2가지 예외처리 후 사용자에게 이해하기 쉬운 메시지 작성 후 출력되도록 코드를 작성

````python
def div_program(arg1, arg2):
    try:
        x = float(arg1)
        y = float(arg2)
        print(f"{x}/{y} = {x/y}", f"{y}/{x} = {y/x}", sep="\n")
    except ValueError:
        print("숫자를 입력하세요")
        return 
    except ZeroDivisionError:
        print("0 이외의 숫자를 입력하세요")
        return 

div_program(2, 1)
````

[문제 2]

파일을 읽어서 내용을 출력하는 프로그램을 작성하세요. 

- 파일이 존재하지 않을 때 발생하는 FileNotFoundError를 처리
- 파일이 존재하지 않을 경우 "파일을 찾을 수 없습니다."라는 메시지를 출력

````python
def file_reader(txt):
    try:
        # str(txt)로 다른 종류의 에러를 최대한 방지
        with open(str(txt), "r") as f:
            print(f.read())
    except FileNotFoundError:
        print("파일을 찾을 수 없습니다")
        return

file_reader("test.txt")
file_reader("notfound.txt")
````

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
def num_list():
    number_list = []
    while True:
        num_input = input("숫자를 입력하세요. 종료하려면 done을 입력하세요: ")

        try:
            if num_input == "done":
                print("종료하였습니다")
                return number_list
            float(num_input)
            number_list.append(float(num_input))
            print(number_list)
        except ValueError:
            print(f"{num_input}은 숫자가 아닙니다")
            continue

num_list()
````

[문제 4]

사용자 정의 예외 NegativeNumberError를 생성하고, 사용자로부터 숫자를 입력받아 음수일 경우 NegativeNumberError를 발생시키는 프로그램을 작성하세요. 

- 예외 처리 구문을 사용하여 음수 입력 시 "양수를 입력해주세요."라는 메시지를 출력

````python
class NegativeNumberError(Exception):
    def __init__(self, message):
        self.error_type = "NegativeNumberError"
        self.error_message = message
    
    def __str__(self):
        return f"{self.error_type}: {self.error_message}"


def negative_generator(*args):
        for i in args:
            try:
                if i < 0:
                    raise NegativeNumberError(f"{i} 말고 양수를 입력해 주세요")
            except NegativeNumberError as e:
                print(e)

negative_generator(1, 2, 3, -1, -2, -3)
````

datetime 모듈을 사용하여 현재 날짜와 시간을 출력하는 프로그램을 작성하세요. 

또한, 사용자로부터 태어난 년도를 입력받아 현재 나이를 계산하여 출력하세요.

````python
import datetime

def now_time():
    print(datetime.datetime.now())

def age(birth_year, birth_month, birth_day):
    today = datetime.date.today()
    korean_age = today.year - birth_year + 1
    
    if (today.month, today.day) > (birth_month, birth_day):
        age = korean_age - 2
    else:
        age = korean_age - 1
    print(f"나이: {korean_age}, 만 나이: {age}")

now_time()
# 사용자: 세종대왕
age(1397, 5, 15)
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def time_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        run_time = end - start
        print(f"함수 실행 시간: {run_time:.2f}초")
    return wrapper

@time_decorator
def long_long_time(*args, **kwargs):
    for i in range(999999999):
        pass
        
long_long_time()
````

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
# mymath.py

def far(ground_area, building_area, floor):
    area_ratio = (floor * building_area / ground_area) * 100
    print(f"용적률: {area_ratio}%")

def dcr(ground_area, building_area):
    coverage_ratio = (building_area / ground_area) * 100
    print(f"건폐율: {coverage_ratio}%")
````

````python
import mymath

mymath.far(300, 200, 3)
mymath.dcr(500, 300)
````
